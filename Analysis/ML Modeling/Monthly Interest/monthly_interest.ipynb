```python
import pandas as pd
```

```python
data_us = pd.read_csv("FRB_H15.csv")
data_us = data_us.drop(range(5))

# Pulling the 6-month sec
data_us = data_us.rename(columns = {"Market yield on U.S. Treasury securities at 6-month   constant maturity, quoted on investment basis":"US MIR % (6mo)",
                                 "Series Description":"Date"})

data_us = data_us[["Date","US MIR % (6mo)"]]
data_us["Date"] = pd.to_datetime(data_us["Date"]).apply(lambda x:x.strftime('%m/%d/%Y'))
print(data_us.describe())
data_us.head(10)
```

```python
data_eu = pd.read_csv("data.csv")
data_eu["Date"] = pd.to_datetime(data_eu["Date"]).apply(lambda x:x.strftime('%m/%d/%Y'))
data_eu = data_eu.rename(columns = {"MIR%":"EU MIR %"})

print(data_eu.describe())
data_eu.head(10)
```

```python
df = pd.merge(data_us, data_eu, on = "Date", how = "left")
print(df.describe())
df.tail()
```

```python
df.to_csv("monthly_interest.csv",index = False)
```
